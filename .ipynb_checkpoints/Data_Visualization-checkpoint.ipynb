The floating graph for data visulization, using animatplot in matplotlib & seaborn.

```python
import cv2
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt
import matplotlib.animation as animation
# import seaborn as sns
```

import model

```python
model = keras.models.load_model("my_model")
true_class_names = ['nothing', 'fist', 'palm', "two fingers up", "shakabra", "o k", "egyptian hand", "U", "L",
                   "talking hand", "V"]
```

import images in .jpg format

```python
base = "Pictures"
if os.path.lexists(os.path.join(os.getcwd(), base)):
    directory = os.listdir(base)

X = []
for item in directory :
    path = base + '/' + item
    files = os.listdir(path)
    for image in files:
        final_path = path + '/' + image
        try:
            img = cv2.imread(final_path)
            img = cv2.resize(img, (160, 120))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            X.append(img)
        except:
            continue
X = np.array( X , dtype = "uint8")
X = X.reshape(len(X), 120, 160, 3)
print ("Images loaded: ", len(X))
plt.imshow( X[2])
```

Generate predictions

```python
output = model.predict(X)
```

```python

print ( true_class_names[np.argmax(output[19])])
```

Draw the chart

```python
%matplotlib notebook
#title of the chart
title = "Gesture Predicted: "
#size of the chart
fig = plt.figure( figsize = (13,7))
#range of data
# plt.xlim(-1,11)
#y is always changing, TODO
# plt.ylim()

# the animation(per frame)
def animate(i):
    plt.cla()
    plt.ylim(0, max(output[i]))
    plt.xlabel("Gestures", fontsize = 20)
    plt.ylabel("Predicted Probabilities", fontsize = 20)
    plt.title(title + true_class_names[np.argmax(output[i])], fontsize = 20)
    clrs = ['blue' if (x == true_class_names[np.argmax(output[i])]) else 'grey' for x in true_class_names ]
    plt.bar(true_class_names, output[i], width = 0.5, color = clrs)
    plt.pause(1)
#     plot = sns.barplot
ani = animation.FuncAnimation(fig, animate, frames = len(X), repeat = False)
plt.show()    
```
